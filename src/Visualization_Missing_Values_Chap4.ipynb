```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/training_filtered.csv')
```

```python
df_columns_info = pd.read_csv('/content/WiDS Datathon 2020 Dictionary.csv')
```

# Lab Measurements

```python
lab_columns = list(df_columns_info[df_columns_info['Category'] == 'labs']['Variable Name'])
df_lab_data = df[lab_columns]
```

```python
missing_percent = df_lab_data.isnull().mean() * 100

missing_percent = missing_percent[missing_percent > 0]

missing_df = missing_percent.reset_index()
missing_df.columns = ['Column', 'MissingPercent']

missing_df = missing_df.sort_values('MissingPercent', ascending=True)

# Plot
plt.figure(figsize=(10, 8))
sns.barplot(data=missing_df, x='MissingPercent', y='Column')
plt.title('Percentage of Missing Values per Column')
plt.xlabel('Missing Values (%)')
plt.ylabel('Column')

plt.axvline(x=70, color='red', linestyle='--')

plt.tight_layout()
plt.show()
```

# APACHE Variables

```python
APACHE_columns = list(df_columns_info[df_columns_info['Category'] == 'APACHE covariate']['Variable Name'])
existing_columns = [col for col in APACHE_columns if col in df.columns]
df_APACHE_data = df[existing_columns]
```

```python
missing_percent = df_APACHE_data.isnull().mean() * 100
missing_percent = missing_percent[missing_percent > 0]

missing_df = missing_percent.reset_index()
missing_df.columns = ['Column', 'MissingPercent']

# Sort for better visualization
missing_df = missing_df.sort_values('MissingPercent', ascending=True)
max_val = missing_df['MissingPercent'].max()
# Plot
plt.figure(figsize=(10, 8))
sns.barplot(data=missing_df, x='MissingPercent', y='Column')
plt.title('Percentage of Missing Values per Column')
plt.xlabel('Missing Values (%)')
plt.ylabel('Column')

# Add a vertical line to emphasize low percentages
plt.axvline(x=max_val, color='red', linestyle='--')

plt.tight_layout()
plt.show()
```
