```python
# List of APACHE features to keep
apache_keep = [
    "apache_3j_bodysystem",
    "apache_2_bodysystem",
    "apache_2_diagnosis",
    "apache_3j_diagnosis",
    "gcs_eyes_apache",
    "gcs_motor_apache",
    "gcs_verbal_apache",
    "gcs_unable_apache",
    "apache_post_operative",
    "arf_apache"
]

# Drop all other apache-related columns
apache_cols = [col for col in df.columns if "apache" in col]
apache_to_drop = [col for col in apache_cols if col not in apache_keep]

df = df.drop(columns=apache_to_drop)
```

```python
# Calculate missing values
missing_values = df.isnull().sum()
missing_df = pd.DataFrame({"Feature": missing_values.index, "Missing_Count": missing_values.values})
missing_df.to_csv("missing_values.csv", index=False)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert to string to match palette keys
df["hospital_death"] = df["hospital_death"].astype(str)

# Count values and calculate percentages
count_values = df["hospital_death"].value_counts().sort_index()
total = len(df)
percentages = (count_values / total * 100).round(2)

# Create the plot
sns.set(style="whitegrid")
plt.figure(figsize=(6, 5))
ax = sns.countplot(data=df, x="hospital_death", hue="hospital_death",
                   palette={"0": "#4c72b0", "1": "#c44e52"}, legend=False)

# Set y-axis limit slightly higher than max for label spacing
y_max = count_values.max() * 1.1
ax.set_ylim(0, y_max)

# Add labels with both count and percentage
for i, (count, pct) in enumerate(zip(count_values, percentages)):
    ax.text(i, count + y_max * 0.02, f"{count} ({pct}%)",
            ha='center', va='bottom', fontsize=11, fontweight='bold')

# Custom labels and title
ax.set_xticklabels(["Lived", "Died"])
plt.title("Hospital Death Outcome Distribution", fontsize=14)
plt.xlabel("Outcome")
plt.ylabel("Number of Patients")
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure hospital_death is treated as categorical
df["hospital_death"] = df["hospital_death"].astype(str)

# Set plot style
sns.set(style="whitegrid")

# Plot
plt.figure(figsize=(8, 5))
ax = sns.histplot(data=df, x="age", hue="hospital_death",
                  multiple="stack", bins=10,
                  palette={"0": "#4c72b0", "1": "#c44e52"}, alpha=0.9)

# Labels and title
ax.set_title("Age Distribution by Hospital Death Outcome", fontsize=14)
ax.set_xlabel("Age")
ax.set_ylabel("Number of Patients")
ax.legend(title="Hospital Death", labels=["Died", "Lived"])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(data=df, x="gender", palette={"M": "#4c72b0", "F": "#c44e52"})

# Add value and percentage annotations
total = len(df)
max_height = 0
for p in ax.patches:
    count = int(p.get_height())
    percentage = 100 * count / total
    ax.annotate(f'{count}\n({percentage:.1f}%)',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')
    if p.get_height() > max_height:
        max_height = p.get_height()

# Increase y-limit to leave space for labels
ax.set_ylim(0, max_height * 1.2)

# Styling
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
sns.despine()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))  # Wider for many categories
ax = sns.countplot(data=df, x="ethnicity", palette="Set3", order=df["ethnicity"].value_counts().index)

# Rotate for readability
plt.xticks(rotation=45, ha="right")

# Add value and percentage annotations
total = len(df)
for p in ax.patches:
    count = int(p.get_height())
    percentage = 100 * count / total
    ax.annotate(f'{count}\n({percentage:.1f}%)',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

# Styling
plt.title("Ethnicity Distribution")
plt.xlabel("Ethnicity")
plt.ylabel("Count")
sns.despine()
plt.tight_layout()
plt.show()
```
